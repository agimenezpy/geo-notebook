```python
from shapely.geometry import Point

point1 = Point(-57.5759, -25.2893)

print(point1)
print(point1.x)
print(point1.y)
point1
```

```python
from shapely.wkt import loads

point2 = loads("POINT (-57.5759 -25.2893)")

point2.buffer(0.1)
```

```python
poly1 = loads("POLYGON ((30 10, 40 40, 20 40, 10 20, 30 10))")

print(poly1.area)
print(poly1.length)
```

```python
point3 = Point(-57.566887, -25.283162)

print(point1.distance(point3))
```
